```python
import numpy as np
from qiskit.visualization import plot_bloch_vector
```

```python
def sph2cart(theta: float, phi: float) -> (float, float, float):
    # theta: polar angle
    # phi: azimuthal angle
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)
    return x, y, z
```

$
\ket{0} = cos \dfrac{0}{2} \ket{0} + e^{i0} sin \dfrac{0}{2} \ket{1}
$

```python
theta = 0
phi = 0
x, y, z = sph2cart(theta, phi)
plot_bloch_vector([x, y, z])
```

$
\ket{1} = cos \dfrac{\pi}{2} \ket{0} + e^{i0} sin \dfrac{\pi}{2} \ket{1}
$

```python
theta = np.pi
phi = 0
x, y, z = sph2cart(theta, phi)
plot_bloch_vector([x, y, z])
```

$
\ket{\psi} = cos \dfrac{\pi/2}{2}\ket{0} + e^{i\pi} sin \dfrac{\pi/2}{2} \ket{1} = \dfrac{\ket{0} - \ket{1}}{\sqrt{2}}
$

```python
theta = np.pi / 2
phi = np.pi
x, y, z = sph2cart(theta, phi)
plot_bloch_vector([x, y, z])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
def indicator_func(x):
    a = np.maximum(x, 0)
    a = a / a
    a = np.nan_to_num(a)
    return a

def expo_dist(x, lmbda):
    ind = indicator_func(x)
    return lmbda * ind * np.exp(lmbda * x * -1)
```

```python
plt.figure(1)
x = np.linspace(-1, 3, 200)
plt.plot(x, expo_dist(x, 1))
```

```python
def laplace(x, mu, gamma):
    return (.5 / gamma) * np.exp(-1 * np.abs(x - mu) / gamma)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.figure()
x = np.linspace(-10, 10, 200)
plt.plot(x, sigmoid(x))
```

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
plt.figure()
x = np.linspace(-10, 10, 300)
plt.plot(x, softplus(x))
```

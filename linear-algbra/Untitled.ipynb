```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.asarray([
    [1, 2],
    [2, 3]
])
```

```python
eig_values, eig_vectors = np.linalg.eig(a)
print(eig_values)
print(eig_vectors)
```

```python
np.zeros(eig_values.shape)
```

```python
plt.close()
plt.figure()
plt.autoscale()
for v in eig_vectors:
    plt.arrow(0, 0, v[0], v[1], head_width=0.05, head_length=0.1)
plt.show()
```

## Random numbers

```python
def norm_prob(x, miu, ro):
    return (1.0 / ro * math.sqrt(2 * math.pi)) * math.exp(-1.0 * (x - miu) * (x - miu) / 2 / ro / ro)
```

```python
xs = range(0, 300)
ps = [norm_prob(x, 150, 50) * 100 for x in xs]
plt.plot(xs, ps)
plt.show()
```

## Harmonic Sinusoids are Orthogonal

```julia
function harmonic(N, k)
    n = 0:N-1
    ω = 2π * k / N
    x = exp.(im * ω .* n)
    return x
end
```

```julia
x = harmonic(40, 0)
stem(n, [real(x) imag(x)], layout=(2,1))
```

```julia
y = harmonic(40, 1)
stem(n, [real(y) imag(y)], layout=(2,1))
```

```julia
z = harmonic(40, 10)
stem(n, [real(z) imag(z)], layout=(2,1))
```

```julia
x' * z
```

## Cauchy Schwartz Inequality

```julia
N = 120
n = 0:N-1

x0 = [zeros(10, 1); ones(100, 1); zeros(10, 1)]
x0 = x0 / norm(x0)
x1 = [zeros(10, 1); ones(50, 1); -ones(50, 1); zeros(10, 1)]
x1 = x1 / norm(x1)
```

```julia
stem(n, x0)
```

```julia
stem(n, x1)
```

```julia
y = x1 + 0.3 * randn(size(x1))
stem(n, y)
```

```julia
abs.(y' * x0), abs.(y' * x1)
```

```julia
norm(x0), norm(x1), norm(y)
```

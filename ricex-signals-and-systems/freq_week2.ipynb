```julia
x = [-1; -2; -4; 2; 3; 5; 2; 0; 1; -4; 1; 3]
n = 0:length(x)-1
stem(n, x, size=(400,200))
```

## Dirichlet Kernel

```julia
M = 3
ω = -π:0.01:π
y = sin.((2M + 1) .* ω ./ 2) ./ sin.(ω ./ 2)
```

```julia
plot(ω, y, size=(600,200))
```

## Recursive Average

```julia
ω = -π:0.01:π
α = 0.9
H = 1 ./ ( 1 - α * e .^ (-im * ω))

plot(ω, abs.(H), size(600,200))
```
